```python
from ultralytics import YOLO
import easyocr
import cv2
import matplotlib.pyplot as plt
```

```python
model = YOLO('yolo11n.pt')
```

```python
model.train(
    data="C:/Users/shahabuddin akhon hr/Desktop/numberPlateDetection\data/license_plate_detect/data.yaml",
    epochs=15,
    batch=10,
    imgsz=640
)
```

```python
model = YOLO('runs/detect/train3/weights/best.pt')
```

```python
def preprocess(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.bilateralFilter(img, 11, 17, 17) #noise detection
#    img = cv2.Canny(img, 30, 200) #edge detection
    plt.figure(figsize=(8, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    return img
```

```python
reader = easyocr.Reader(['en'])
def perform_ocr_on_image(img,coordinates):
    x_min, y_min, x_max, y_max = map(int, coordinates)
    width = x_max - x_min
    height = y_max - y_min
    cropped_img = img[y_min:y_min+height, x_min:x_min+width]
    preprocessed_img = preprocess(cropped_img)
    data = reader.readtext(preprocessed_img)
    text = ""
    for letter in data:
        if len(letter) == 1 or (len(letter[1]) > 6 and letter[2] > 0.2):
            text = letter[1]
    return str(text)
```

```python
def extract_text_from_yolo_result(result):
    orig_img = result.orig_img
    boxes = result.boxes.xyxy
    if len(boxes) > 0:
        license_plate_box = boxes[0].cpu().numpy()
        
        detected_text = perform_ocr_on_image(orig_img, license_plate_box)
        print(f"Detected license plate text: {detected_text}")
    else:
        print("No license plate detected.")
```

```python
result = model.predict('number_plate.jpg')
result[0].save()
extract_text_from_yolo_result(result[0])
```

```python
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error opening video file")
        return

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model.predict(frame, conf=0.3)

        first_result = results[0]
        boxes = first_result.boxes.xyxy

        for box in boxes:
            license_plate_box = box.cpu().numpy()
            print(f"Bounding box coordinates: {box.cpu().numpy()}")
            detected_text = perform_ocr_on_image(frame, license_plate_box)
            print(f"Detected license plate text: {detected_text}")
            x_min, y_min, x_max, y_max = map(int, license_plate_box)
            cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 255), 2)
            text_position = (x_min, y_min - 10)
            (text_width, text_height), baseline = cv2.getTextSize(detected_text, cv2.FONT_HERSHEY_SIMPLEX, 0.7, 2)
            cv2.rectangle(frame, (x_min, y_min - text_height - 10), (x_min + text_width, y_min), (255, 0, 0), cv2.FILLED)
            cv2.putText(frame, detected_text, text_position, cv2.FONT_HERSHEY_SIMPLEX, 
                        0.7, (255, 255, 255), 2, cv2.LINE_AA)

        cv2.imshow('License Plate Detection', frame)
        
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
process_video('data/car.mp4')
```
